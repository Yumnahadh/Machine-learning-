# 1. My lane (or freestyle) and why

**Lane Choice:** Lane 1 — Page-level Content Refresh Prioritization.

**Why this lane?** 
In a large content ecosystem (~30k+ pages), editorial teams cannot manually audit every URL each month to identify decay. I am picking Lane 1 because it directly targets a high-leverage business problem: identifying which decaying pages have the highest recovery potential before organic traffic and conversions are permanently lost.

# 2. The question: decision, action, cost of a wrong call

* **The Decision:** Deciding which published pages should be assigned to the content/editorial team for a manual content refresh in the next sprint.
* **Who Acts On It:** Content managers, SEO strategists, and staff writers.
* **The Action:** An editor updates outdated facts, fills content gaps, and republishes the page to regain lost search visibility.
* **Unit of Analysis:** A single URL / page (`page_id`).
* **Cost of False Positive:** Wastes precious editorial hours rewriting a healthy page, delaying updates on truly decaying pages.
* **Cost of False Negative:** Continued loss of organic search traffic, conversions, and rankings to competitors.

```python
import pandas as pd

# Load dataset from starter repository directly
url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

total_pages = len(df)
decayed_pages = df['label_refresh_needed'].sum() if 'label_refresh_needed' in df.columns else (df['impressions_drop_ratio'] > 0.3).sum()
decay_pct = (decayed_pages / total_pages) * 100
avg_impression_loss = df[df['label_refresh_needed'] == 1]['impressions_30d_delta'].mean() if 'label_refresh_needed' in df.columns else df['impressions_30d_delta'].mean()

print("=== STARTER DATASET SUMMARY METRICS ===")
print(f"1. Total Dataset Volume: {total_pages:,} total page records evaluated.")
print(f"2. Baseline Refresh Opportunity: {decayed_pages:,} pages ({decay_pct:.2f}%) are classified as needing a refresh.")
print(f"3. Observed Delta Impact: Flagged pages show an average 30-day loss of {avg_impression_loss:.1f} impressions.")
```
